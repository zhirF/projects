```python
import re, string, numpy as np, pandas as pd, matplotlib.pyplot as plt
from pathlib import Path
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

data_dir = Path("archive")
files = sorted([p for p in data_dir.rglob("*.txt")])

texts_raw, labels = [], []
for p in files:
    try:
        txt = p.read_text(encoding="utf-8", errors="ignore")
    except:
        txt = p.read_text(encoding="latin-1", errors="ignore")
    texts_raw.append(txt)
    labels.append(p.parent.name)

stop = set(ENGLISH_STOP_WORDS)
punct = str.maketrans("", "", string.punctuation)

def clean_tokens(s):
    s = s.lower().translate(punct)
    s = re.sub(r"\d+", " ", s)
    toks = s.split()
    toks = [t for t in toks if len(t) > 2 and t not in stop]
    return toks

token_lists = [clean_tokens(t) for t in texts_raw]
docs = [" ".join(toks) for toks in token_lists]
labels = np.array(labels)

pd.DataFrame({"label": labels}).value_counts().to_frame("n").head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

cv = CountVectorizer(max_df=0.5, min_df=5, ngram_range=(1,1))
X_counts = cv.fit_transform(docs)

tfidf = TfidfVectorizer(max_df=0.5, min_df=5, ngram_range=(1,1))
X_tfidf = tfidf.fit_transform(docs)

X_counts.shape, X_tfidf.shape
```

```python
from sklearn.decomposition import LatentDirichletAllocation
np.random.seed(42)

n_topics = 10
lda = LatentDirichletAllocation(
    n_components=n_topics,
    learning_method="batch",
    max_iter=20,
    random_state=42
).fit(X_counts)

perplexity = lda.perplexity(X_counts)
feature_names = np.array(cv.get_feature_names_out())

topn = min(15, X_counts.shape[1])
topic_words = []
topic_weights = []
for k in range(n_topics):
    idx = np.argsort(lda.components_[k])[-topn:][::-1]
    topic_words.append([feature_names[i] for i in idx])
    topic_weights.append(lda.components_[k][idx])

pd.DataFrame({
    "topic": np.arange(n_topics),
    "top_words": [", ".join(w) for w in topic_words]
}), perplexity
```

```python
from gensim.corpora.dictionary import Dictionary
from gensim.models.coherencemodel import CoherenceModel

dictionary = Dictionary(token_lists)
cm = CoherenceModel(topics=topic_words, texts=token_lists, dictionary=dictionary, coherence="c_v")
coherence = cm.get_coherence()

doc_topic = lda.transform(X_counts)
dominant_topic = doc_topic.argmax(axis=1)

summary_df = pd.DataFrame({
    "label": labels,
    "dominant_topic": dominant_topic
})
topic_counts = summary_df["dominant_topic"].value_counts().sort_index()
summary_df.head(), coherence, topic_counts.to_frame("docs")
```

```python
max_show = min(6, len(topic_words))
for k in range(max_show):
    words = topic_words[k]
    weights = topic_weights[k]
    plt.figure(figsize=(10,5))
    plt.bar(range(len(weights)), weights)
    plt.xticks(range(len(words)), words, rotation=45, ha="right")
    plt.title(f"Topic {k}: Top Words")
    plt.tight_layout()
    plt.show()

# documents per dominant topic
counts = pd.Series(dominant_topic).value_counts().sort_index()
plt.figure(figsize=(8,4))
plt.bar(counts.index, counts.values)
plt.xlabel("Dominant Topic")
plt.ylabel("Document Count")
plt.title("Documents per Dominant Topic")
plt.tight_layout()
plt.show()

try:
    from bertopic import BERTopic
    topic_model = BERTopic(verbose=False, random_state=42)
    topics, probs = topic_model.fit_transform(docs)
    topic_model.get_topic_info().head(12)
except Exception as e:
    str(e)
```
